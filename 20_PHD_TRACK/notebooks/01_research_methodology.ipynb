## Part 1: How to Read Research Papers

### The Three-Pass Method (S. Keshav)

#### Pass 1: Bird's Eye View (5-10 minutes)

**Goal**: Understand if the paper is relevant.

**What to read**:
1. ✅ **Title, Abstract, Introduction**
2. ✅ **Section headings** (don't read content yet)
3. ✅ **Conclusion**
4. ✅ **Figures and tables** (what are they showing?)
5. ✅ **References** (recognize any? foundational work?)

**Questions to answer**:
- What category? (Theory, system, empirical?)
- What's the main contribution?
- Is it well-written?
- Worth reading further?

**Decision**: 
- ❌ Not relevant → Stop
- ✅ Interesting → Continue to Pass 2
- 🔥 Core to your work → Continue to Pass 3

---

#### Pass 2: Grasp with Care (1 hour)

**Goal**: Understand the paper's content (but not deeply).

**What to do**:
1. Read **carefully** but skip proofs
2. **Annotate** key points, insights, unclear parts
3. **Examine figures and diagrams** - understand every axis, legend, curve
4. **Mark unread references** that seem important

**Questions to answer**:
- What's the problem they're solving?
- What's their approach/method?
- What are the key results?
- What are the assumptions?
- Can I explain this to someone else?

**Output**: 
- Summary in your own words (1 paragraph)
- List of strengths and weaknesses

---

#### Pass 3: Virtual Re-implementation (4-5 hours)

**Goal**: Deeply understand every detail.

**What to do**:
1. **Re-implement mentally** - imagine implementing every algorithm
2. **Challenge every assumption** - why is this needed?
3. **Identify implicit assumptions** not stated
4. **Question every claim** - does the evidence support it?
5. **Compare with related work** - what's truly novel?
6. **Brainstorm improvements** - what would you do differently?

**Questions to answer**:
- What are the innovation points?
- What are the technical strengths?
- What are the weaknesses/limitations?
- What future work does this enable?
- Can I reproduce this?

**Output**:
- Detailed notes with diagrams
- List of questions for authors
- Ideas for follow-up work

### Practical Exercise: Reading "Attention Is All You Need"

Let's apply the three-pass method to the Transformer paper.

#### Pass 1 (10 min)

**Title**: "Attention Is All You Need" (Vaswani et al., 2017)

**Abstract Summary**:
- Proposes Transformer: new architecture based solely on attention
- No recurrence, no convolution
- SOTA on translation tasks
- Faster to train than RNN/CNN

**Section Headings**:
1. Introduction
2. Background
3. Model Architecture
   - 3.1 Encoder-Decoder
   - 3.2 Attention
   - 3.3 Position-wise FFN
   - 3.4 Embeddings
   - 3.5 Positional Encoding
4. Why Self-Attention
5. Training
6. Results
7. Conclusion

**Key Figures**:
- Figure 1: Overall architecture (encoder-decoder stack)
- Figure 2: Scaled Dot-Product Attention & Multi-Head Attention

**Main Contribution**: Self-attention mechanism can replace RNNs for sequence transduction.

**Decision**: 🔥 Foundational work → Pass 3 needed

---

#### Pass 2 (1 hour)

**Problem**: 
- RNNs are sequential → can't parallelize → slow training
- Long-range dependencies still challenging despite LSTM/GRU

**Method**:
- Self-attention: Directly model dependencies regardless of distance
- Multi-head: Attend to different representation subspaces
- Positional encoding: Inject sequence order information

**Key Innovation**:
$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

Scaling by $\sqrt{d_k}$ prevents softmax saturation.

**Results**:
- WMT 2014 EN→DE: 28.4 BLEU (new SOTA)
- WMT 2014 EN→FR: 41.8 BLEU (new SOTA)
- Trains 10x faster than previous best model

**Strengths**:
- ✅ Parallelizable (no sequential dependencies)
- ✅ Constant path length between any two positions
- ✅ More interpretable (attention weights)

**Weaknesses**:
- ❌ Quadratic complexity $O(n^2 \cdot d)$ in sequence length
- ❌ Needs positional encoding (not learned inherently)
- ❌ Memory intensive for long sequences

---

#### Pass 3 (4-5 hours)

**Deep Dive Questions**:

1. **Why $\sqrt{d_k}$ scaling?**
   - As $d_k$ grows, dot products grow large → softmax saturates
   - Standard deviation of dot product is $\sqrt{d_k}$ (assuming unit variance)
   - Scaling keeps values in reasonable range

2. **Why sinusoidal positional encoding?**
   - Allows extrapolation to longer sequences
   - $PE_{pos, 2i} = \sin(pos / 10000^{2i/d})$
   - Different frequencies for different dimensions
   - Model can learn relative positions via linear combinations

3. **Why multi-head attention?**
   - Single attention focuses on one aspect
   - Multiple heads capture different relationships
   - Empirically: 8 heads works well

4. **What's the computational complexity vs RNN?**
   - Self-attention: $O(n^2 \cdot d)$ per layer
   - RNN: $O(n \cdot d^2)$ per layer
   - For $n < d$, self-attention is faster (typical case)

**Implementation Details** (from code):
- 6 encoder layers, 6 decoder layers
- $d_{model} = 512$, $d_{ff} = 2048$
- 8 attention heads ($d_k = d_v = 64$)
- Dropout = 0.1
- Label smoothing = 0.1
- Warmup + inverse sqrt schedule

**Ablation Studies** (Table 3):
- Removing positional encoding: -1.2 BLEU
- Different attention types tested
- Varying # of heads: 8 is optimal

**Ideas for Extensions**:
1. Reduce $O(n^2)$ complexity → Led to Longformer, BigBird, etc.
2. Better positional encodings → RoPE, ALiBi
3. Decoder-only for LLMs → GPT
4. Encoder-only for embeddings → BERT

## Part 3: Implementing Papers from Scratch

### Step-by-Step Process

#### Step 1: Understand the Algorithm

Before coding:
1. Draw the architecture on paper
2. Write pseudocode
3. Identify all hyperparameters
4. Understand data flow (tensor shapes)

#### Step 2: Start Simple

**Don't** implement everything at once!

**Example**: Implementing Transformer
```
Day 1: Scaled dot-product attention (test on small tensors)
Day 2: Multi-head attention (verify shape transformations)
Day 3: Single encoder layer (test with dummy data)
Day 4: Positional encoding (visualize embeddings)
Day 5: Full encoder stack
Day 6: Decoder (causal masking)
Day 7: Training loop
Day 8-10: Debug, optimize, validate
```

#### Step 3: Test Each Component

Write unit tests for EVERY component.

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

# Example: Testing attention mechanism
def test_attention_output_shape():
    """Test 1: Output shape should match input shape."""
    batch, seq_len, d_model = 2, 10, 64
    Q = K = V = torch.randn(batch, seq_len, d_model)
    
    def scaled_dot_product_attention(Q, K, V):
        d_k = Q.size(-1)
        scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(d_k)
        attn_weights = F.softmax(scores, dim=-1)
        output = torch.matmul(attn_weights, V)
        return output, attn_weights
    
    output, weights = scaled_dot_product_attention(Q, K, V)
    
    assert output.shape == (batch, seq_len, d_model), f"Expected {(batch, seq_len, d_model)}, got {output.shape}"
    assert weights.shape == (batch, seq_len, seq_len), f"Expected {(batch, seq_len, seq_len)}, got {weights.shape}"
    print("✅ Test 1 passed: Output shapes correct")

def test_attention_weights_sum_to_one():
    """Test 2: Attention weights should sum to 1 (valid probability distribution)."""
    batch, seq_len, d_model = 2, 10, 64
    Q = K = V = torch.randn(batch, seq_len, d_model)
    
    def scaled_dot_product_attention(Q, K, V):
        d_k = Q.size(-1)
        scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(d_k)
        attn_weights = F.softmax(scores, dim=-1)
        output = torch.matmul(attn_weights, V)
        return output, attn_weights
    
    output, weights = scaled_dot_product_attention(Q, K, V)
    
    # Sum over key dimension should be 1
    sums = weights.sum(dim=-1)
    assert torch.allclose(sums, torch.ones_like(sums), atol=1e-5), "Attention weights don't sum to 1!"
    print("✅ Test 2 passed: Attention weights sum to 1")

def test_causal_mask():
    """Test 3: Causal mask should prevent attending to future tokens."""
    seq_len = 5
    d_model = 8
    Q = K = V = torch.randn(1, seq_len, d_model)
    
    # Create causal mask
    mask = torch.tril(torch.ones(seq_len, seq_len)).bool()
    
    def masked_attention(Q, K, V, mask):
        d_k = Q.size(-1)
        scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(d_k)
        scores = scores.masked_fill(~mask, float('-inf'))
        attn_weights = F.softmax(scores, dim=-1)
        output = torch.matmul(attn_weights, V)
        return output, attn_weights
    
    output, weights = masked_attention(Q, K, V, mask)
    
    # Upper triangle should be zero (can't attend to future)
    upper_triangle = torch.triu(weights[0], diagonal=1)
    assert torch.allclose(upper_triangle, torch.zeros_like(upper_triangle), atol=1e-5), "Causal mask not working!"
    print("✅ Test 3 passed: Causal mask prevents future attention")

# Run tests
test_attention_output_shape()
test_attention_weights_sum_to_one()
test_causal_mask()

print("\n✅ All tests passed! Implementation is correct.")
````

#### Step 4: Reproduce Paper Results

**Goal**: Match the paper's reported metrics.

**Common reasons for mismatch**:
1. ❌ Different initialization
2. ❌ Missing normalization layers
3. ❌ Wrong learning rate schedule
4. ❌ Data preprocessing differences
5. ❌ Batch size too small (affects BatchNorm)
6. ❌ Missing regularization (dropout, weight decay)
7. ❌ Random seeds (some papers don't report)

**Debugging Strategy**:
```
1. Start with smaller model/dataset (faster iteration)
2. Compare intermediate outputs with reference implementation
3. Check gradient flow (vanishing/exploding)
4. Verify loss is decreasing
5. Use same exact hyperparameters as paper
6. If paper has code, compare line-by-line
```

**Sanity Checks**:

````python
import matplotlib.pyplot as plt
import numpy as np

def sanity_check_model(model, dataloader):
    """
    Run sanity checks before full training.
    """
    print("Running sanity checks...\n")
    
    # Check 1: Model can overfit single batch
    print("1. Testing overfitting on single batch...")
    model.train()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
    batch = next(iter(dataloader))
    x, y = batch
    
    losses = []
    for i in range(100):
        optimizer.zero_grad()
        output = model(x)
        loss = F.cross_entropy(output, y)
        loss.backward()
        optimizer.step()
        losses.append(loss.item())
    
    if losses[-1] < 0.1 * losses[0]:
        print(f"   ✅ Loss decreased: {losses[0]:.4f} → {losses[-1]:.4f}")
    else:
        print(f"   ❌ Model not learning on single batch!")
        print(f"   Loss: {losses[0]:.4f} → {losses[-1]:.4f}")
    
    # Check 2: Gradients are flowing
    print("\n2. Checking gradient flow...")
    model.zero_grad()
    output = model(x)
    loss = F.cross_entropy(output, y)
    loss.backward()
    
    gradient_norms = []
    for name, param in model.named_parameters():
        if param.grad is not None:
            gradient_norms.append(param.grad.norm().item())
        else:
            print(f"   ❌ No gradient for {name}")
    
    if len(gradient_norms) > 0:
        avg_grad = np.mean(gradient_norms)
        if avg_grad > 1e-7:
            print(f"   ✅ Gradients flowing (avg norm: {avg_grad:.6f})")
        else:
            print(f"   ⚠️  Gradients very small (avg norm: {avg_grad:.6f})")
    
    # Check 3: Output distribution
    print("\n3. Checking output distribution...")
    model.eval()
    with torch.no_grad():
        outputs = []
        for batch in dataloader:
            x, _ = batch
            output = model(x)
            outputs.append(output)
            if len(outputs) >= 10:  # Check a few batches
                break
        outputs = torch.cat(outputs)
    
    if torch.isnan(outputs).any():
        print("   ❌ NaN in outputs!")
    elif torch.isinf(outputs).any():
        print("   ❌ Inf in outputs!")
    else:
        print(f"   ✅ Outputs finite (mean: {outputs.mean():.4f}, std: {outputs.std():.4f})")
    
    print("\nSanity checks complete!")

# Example usage:
# sanity_check_model(your_model, train_dataloader)
````

## Part 4: Conducting Ablation Studies

### What is an Ablation Study?

**Definition**: Systematically remove/change components to understand their contribution.

**Goal**: Answer "What is each component worth?"

### Design Principles

1. **Baseline**: Full model (all components)
2. **Ablations**: Remove one component at a time
3. **Comparison**: Measure performance drop
4. **Analysis**: Understand why each component helps

### Example: Transformer Ablations

| Variant | BLEU | Δ from baseline |
|---------|------|------------------|
| **Full model** | 28.4 | - |
| - Positional encoding | 27.2 | -1.2 |
| - Multi-head (use single head) | 26.5 | -1.9 |
| - Residual connections | 25.1 | -3.3 |
| - Layer normalization | 24.8 | -3.6 |
| - Attention (use FFN only) | 18.2 | -10.2 |

**Interpretation**:
- Attention is most critical (-10.2 BLEU)
- LayerNorm and residuals important for training stability
- Multi-head provides significant benefit over single head

### Hyperparameter Ablations

| Hyperparameter | Values tested | Best |
|----------------|---------------|------|
| # of heads | 1, 2, 4, 8, 16 | 8 |
| $d_{model}$ | 256, 512, 1024 | 512 |
| # layers | 2, 4, 6, 8 | 6 |
| Dropout | 0, 0.1, 0.2, 0.3 | 0.1 |

### Code for Ablation Studies

````python
import pandas as pd

def run_ablation_study(base_config, ablations, train_fn, eval_fn):
    """
    Run systematic ablation study.
    
    Args:
        base_config: dict with all hyperparameters
        ablations: dict mapping name → config changes
        train_fn: function(config) → model
        eval_fn: function(model) → metric
    
    Returns:
        pd.DataFrame with results
    """
    results = []
    
    # Baseline
    print("Training baseline...")
    baseline_model = train_fn(base_config)
    baseline_metric = eval_fn(baseline_model)
    results.append({
        'variant': 'Baseline (full model)',
        'metric': baseline_metric,
        'delta': 0.0
    })
    print(f"Baseline: {baseline_metric:.4f}\n")
    
    # Ablations
    for name, config_changes in ablations.items():
        print(f"Training ablation: {name}...")
        
        # Merge changes with base config
        ablation_config = {**base_config, **config_changes}
        
        # Train and evaluate
        model = train_fn(ablation_config)
        metric = eval_fn(model)
        delta = metric - baseline_metric
        
        results.append({
            'variant': name,
            'metric': metric,
            'delta': delta
        })
        print(f"  Metric: {metric:.4f} (Δ = {delta:+.4f})\n")
    
    # Create DataFrame
    df = pd.DataFrame(results)
    df = df.sort_values('delta', ascending=True)  # Sort by impact
    
    return df

# Example usage
base_config = {
    'n_layers': 6,
    'n_heads': 8,
    'd_model': 512,
    'dropout': 0.1,
    'use_positional_encoding': True,
    'use_residual': True,
    'use_layer_norm': True
}

ablations = {
    'No positional encoding': {'use_positional_encoding': False},
    'Single head': {'n_heads': 1},
    'No residual connections': {'use_residual': False},
    'No layer norm': {'use_layer_norm': False},
    'Fewer layers (4)': {'n_layers': 4},
    'Smaller model (256)': {'d_model': 256},
    'No dropout': {'dropout': 0.0}
}

# results_df = run_ablation_study(base_config, ablations, train_model, evaluate_model)
# print(results_df)

print("Template for ablation studies created!")
````

## Part 5: Writing Research Papers

### Paper Structure (Standard)

#### 1. Title
- **Descriptive** but **concise**
- Contains key contribution
- Good: "Attention Is All You Need"
- Bad: "A Novel Deep Learning Approach" (too vague)

#### 2. Abstract (150-250 words)

**Template** (4 sentences):
```
1. [Problem]: What problem are you solving?
2. [Method]: What's your approach?
3. [Results]: What did you achieve?
4. [Impact]: Why does it matter?
```

**Example** (GPT-3 paper):
> *Recent work has demonstrated substantial gains on many NLP tasks through pre-training large language models on a diverse corpus. We show that scaling up language models greatly improves task-agnostic, few-shot performance. Specifically, we train GPT-3, a 175 billion parameter autoregressive model and test its performance in the few-shot setting. We find that GPT-3 achieves strong performance on many NLP datasets, including translation, question-answering, and cloze tasks.*

#### 3. Introduction

**Structure** (1-2 pages):
1. **Motivation**: Why is this problem important?
2. **Limitations of prior work**: What's missing?
3. **Your contribution**: What do you propose?
4. **Key results**: Summary of empirical findings
5. **Paper organization**: "The rest of the paper is organized as follows..."

#### 4. Related Work

- Group by themes (not chronological)
- Explain how your work differs
- Be **generous** but **clear** about differences

#### 5. Method

**Goal**: Reader should be able to reproduce your work.

- Start with intuition, then formalism
- Use figures liberally
- Define all notation
- Walk through algorithm step-by-step

#### 6. Experiments

**Structure**:
1. **Setup**: Datasets, baselines, metrics, hyperparameters
2. **Main results**: Tables comparing to baselines
3. **Ablation studies**: What components matter?
4. **Analysis**: Error analysis, qualitative examples

**Tables**:
- Bold the best results
- Include standard deviations if stochastic
- Use same metrics as prior work for comparison

#### 7. Discussion

- What did we learn?
- Limitations of your approach
- Broader impact (ethical considerations)

#### 8. Conclusion

- Restate contribution (different words than abstract)
- Future work

---

### Writing Tips

1. **Clarity over cleverness**: Simple, direct language
2. **Active voice**: "We propose" not "It is proposed"
3. **One idea per sentence**: Break up complex sentences
4. **Figures tell the story**: Paper should be understandable from figures alone
5. **Iterate**: Write → Get feedback → Revise (10+ times)

### Common Mistakes

❌ **Overselling**: "Revolutionary", "unprecedented"
❌ **Underselling**: "We propose a simple method" (let results speak)
❌ **Missing ablations**: Reviewers will ask "What if you removed X?"
❌ **Cherry-picking results**: Report all experiments, even failures
❌ **Poor notation**: Define symbols, be consistent
❌ **No error bars**: Always report variance
