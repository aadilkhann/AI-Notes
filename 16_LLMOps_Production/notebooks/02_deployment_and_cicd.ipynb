## 1. CI/CD Pipeline for LLMs

**Traditional ML vs LLM CI/CD**:

```
Traditional ML:
Code → Train → Evaluate → Deploy

LLM Pipeline:
Prompt → Test → Eval → Deploy
   ↓       ↓      ↓       ↓
Version  Unit   Metrics  Monitor
Control  Tests  (qual)   Cost
```

**LLM CI/CD Stages**:

1. **Development**:
   - Prompt engineering
   - Local testing
   - Version control (Git)

2. **Testing**:
   - Unit tests (example inputs)
   - Regression tests (golden datasets)
   - Performance tests (latency)
   - Safety tests (content filtering)

3. **Staging**:
   - Deploy to staging environment
   - Run integration tests
   - Load testing
   - Manual QA

4. **Production**:
   - Canary deployment (5% traffic)
   - Monitor metrics
   - Gradual rollout (25% → 50% → 100%)
   - Rollback if issues

## 2. Deployment Strategies

**Deployment Patterns**:

### 1. Blue-Green Deployment
- Two identical environments (Blue = current, Green = new)
- Deploy new version to Green
- Switch traffic from Blue to Green
- Keep Blue for instant rollback

### 2. Canary Deployment
- Deploy new version to small subset (5%)
- Monitor metrics closely
- Gradually increase traffic
- Rollback if issues detected

### 3. Rolling Deployment
- Update instances one by one
- No downtime
- Slower rollout
- Can't easily rollback mid-deployment

### 4. Shadow Deployment
- Route traffic to both old and new
- Serve old version to users
- Log new version outputs for comparison
- No user impact

## 3. A/B Testing

**A/B Testing for LLMs**:

Test different:
- Prompts
- Models (GPT-4 vs Claude)
- Parameters (temperature, top_p)
- RAG strategies
- System prompts

**Process**:
1. **Split traffic**: 50% variant A, 50% variant B
2. **Collect metrics**: Latency, cost, quality, user feedback
3. **Statistical test**: Is B significantly better?
4. **Deploy winner**: Roll out winning variant

**Key Metrics**:
- **Primary**: User satisfaction, task completion
- **Secondary**: Latency, cost, safety
- **Sample size**: Need 1000+ samples for significance

````python
@dataclass
class ABTestResult:
    """A/B test results."""
    variant: str
    n_samples: int
    mean_latency_ms: float
    mean_cost_usd: float
    mean_rating: float  # 1-5
    success_rate: float  # 0-1

def simulate_ab_test(n_samples: int = 1000) -> Tuple[ABTestResult, ABTestResult]:
    """
    Simulate A/B test between two variants.
    
    Variant A: GPT-4 (slower, more expensive, higher quality)
    Variant B: GPT-3.5 (faster, cheaper, slightly lower quality)
    """
    
    # Simulate Variant A (GPT-4)
    a_latencies = np.random.gamma(shape=2, scale=150, size=n_samples)  # mean ~300ms
    a_costs = a_latencies / 1000 * 0.03  # $0.03 per 1K tokens
    a_ratings = np.clip(np.random.normal(4.2, 0.6, size=n_samples), 1, 5)
    a_success = (np.random.random(n_samples) > 0.02).astype(float)  # 98% success
    
    variant_a = ABTestResult(
        variant="A (GPT-4)",
        n_samples=n_samples,
        mean_latency_ms=a_latencies.mean(),
        mean_cost_usd=a_costs.mean(),
        mean_rating=a_ratings.mean(),
        success_rate=a_success.mean()
    )
    
    # Simulate Variant B (GPT-3.5)
    b_latencies = np.random.gamma(shape=2, scale=100, size=n_samples)  # mean ~200ms
    b_costs = b_latencies / 1000 * 0.002  # $0.002 per 1K tokens (15x cheaper)
    b_ratings = np.clip(np.random.normal(3.8, 0.7, size=n_samples), 1, 5)
    b_success = (np.random.random(n_samples) > 0.05).astype(float)  # 95% success
    
    variant_b = ABTestResult(
        variant="B (GPT-3.5)",
        n_samples=n_samples,
        mean_latency_ms=b_latencies.mean(),
        mean_cost_usd=b_costs.mean(),
        mean_rating=b_ratings.mean(),
        success_rate=b_success.mean()
    )
    
    return variant_a, variant_b, a_ratings, b_ratings, a_latencies, b_latencies

# Run A/B test
variant_a, variant_b, a_ratings, b_ratings, a_latencies, b_latencies = simulate_ab_test(1000)

print("\nA/B Test Results:")
print("="*70)

print(f"\n{variant_a.variant}:")
print(f"  Latency: {variant_a.mean_latency_ms:.1f} ms")
print(f"  Cost: ${variant_a.mean_cost_usd:.4f} per request")
print(f"  Rating: {variant_a.mean_rating:.2f}/5.00")
print(f"  Success Rate: {variant_a.success_rate*100:.1f}%")

print(f"\n{variant_b.variant}:")
print(f"  Latency: {variant_b.mean_latency_ms:.1f} ms")
print(f"  Cost: ${variant_b.mean_cost_usd:.4f} per request")
print(f"  Rating: {variant_b.mean_rating:.2f}/5.00")
print(f"  Success Rate: {variant_b.success_rate*100:.1f}%")

# Statistical significance test (t-test)
from scipy import stats

t_stat, p_value = stats.ttest_ind(a_ratings, b_ratings)

print(f"\n\nStatistical Significance (Rating):")
print(f"  t-statistic: {t_stat:.3f}")
print(f"  p-value: {p_value:.4f}")
if p_value < 0.05:
    print(f"  ✓ Significant difference (p < 0.05)")
    winner = "A" if variant_a.mean_rating > variant_b.mean_rating else "B"
    print(f"  → Winner: Variant {winner}")
else:
    print(f"  ⚠ No significant difference (p >= 0.05)")

# Decision
print(f"\n\nDecision Analysis:")
print(f"  Variant A: {variant_a.mean_rating - variant_b.mean_rating:.2f} better rating")
print(f"  But {variant_a.mean_latency_ms / variant_b.mean_latency_ms:.1f}x slower")
print(f"  And {variant_a.mean_cost_usd / variant_b.mean_cost_usd:.0f}x more expensive")
print(f"\n  Recommendation:")
if variant_a.mean_rating - variant_b.mean_rating > 0.3:
    print(f"    → Use Variant A (quality worth the cost)")
else:
    print(f"    → Use Variant B (cost savings outweigh quality difference)")
````

   - **Canary**: Low risk, gradual rollout
   - **Rolling**: No downtime, minor updates
   - **Shadow**: Zero risk, major changes

3. **A/B Testing**:
   - Need 1000+ samples for significance
   - Test one variable at a time
   - Consider cost-quality tradeoffs
   - Use statistical tests (t-test, chi-square)

4. **Production Checklist**:
   - ✓ Automated testing (unit, integration, E2E)
   - ✓ Staging environment
   - ✓ Gradual rollout (5% → 25% → 50% → 100%)
   - ✓ Monitoring dashboards
   - ✓ Alerting on key metrics
   - ✓ Rollback plan (<5 min)
   - ✓ Cost tracking
   - ✓ Safety/moderation

5. **Best Practices**:
   - Treat prompts as code (version control)
   - Automate everything
   - Monitor continuously
   - Deploy frequently (daily/weekly)
   - Test  in production (shadow/canary)
   - Fast rollback (automatic)
   - Post-mortems for incidents
